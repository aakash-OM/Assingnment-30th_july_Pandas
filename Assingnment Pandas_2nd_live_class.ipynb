```python
import pandas as pd
```

```python
df = pd.read_csv('AgentPerformance.csv')
df
```

```python
# Q1. Find out there avarage rating on weekly basis keep this in a mind that they take two days of leave.
```

```python
df.dtypes
```

```python
'''first convert 'Date' col into datetime type'''
```

```python
df['Date']=pd.to_datetime(df['Date'])
```

```python
df.dtypes
```

```python
df['cnvt_date_week'] = df['Date'].dt.week
```

```python
df.groupby('cnvt_date_week')['Average Rating'].mean()
```

```python
# Q2. Total working days for each agents.
```

```python
df['cnvt_date_days'] = df['Date'].dt.day
```

```python
df[['Agent Name','cnvt_date_days']]
```

```python
'''Agent names working maximum '''

df[df['cnvt_date_days'] == max(df['cnvt_date_days'])]
```

```python
'''Agent names working minimum days '''

df[df['cnvt_date_days'] == min(df['cnvt_date_days'])]
```

```python
# Q3. Total query that you hvae taken.
```

```python
'''Count the chats col(queries) '''
df[['Agent Name','Total Chats']]
```

```python
'''Maximum chats taken by Agents'''

df[df['Total Chats']==max(df['Total Chats'])]
```

```python
'''Minimum chats taken by Agents'''

df[df['Total Chats']==min(df['Total Chats'])]
```
